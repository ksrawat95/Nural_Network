```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x1 = np.random.randint(3, 12, 10)
x2 = np.random.randint(9, 18, 10)
x3 = np.random.randint(12, 20, 10)
y = x1*7 + 5*x2 + 4*x3
```

```python
x1,x2,x3
```

```python
SHAPE = x1.shape[0]
SHAPE
```

```python
y
```

```python
y = y.reshape(SHAPE,1)
y
```

```python
X = np.array([x1, x2, x3])
X
```

```python
X = X.T
X
```

```python
lr = 0.0001
w = np.array([np.random.randn(1),
              np.random.randn(1),
              np.random.randn(1)])
w
```

```python
error_list = []
for i in range(50):
    y_pred = np.matmul(X, w)
    error = (y - y_pred)**2
    dEw1 = 2*np.matmul((y - y_pred).T, (-X[:,0].reshape(SHAPE, 1)))
    dEw2 = 2*np.matmul((y - y_pred).T, (-X[:,1].reshape(SHAPE, 1)))
    dEw3 = 2*np.matmul((y - y_pred).T, (-X[:,2].reshape(SHAPE, 1)))
    
    w[0][0] = w[0][0] - lr * dEw1
    w[1][0] = w[1][0] - lr * dEw2
    w[2][0] = w[2][0] - lr * dEw3
    
    error_list.append(error.mean())
    print("value of w1 - ", w[0][0], "value of w2 - ", w[1][0], "value of w3 - ", w[2][0],error)
   
```

```python
 plt.plot(error_list)
```
